# n ^ m 구하기
옵션
임의의 정수 n, m이 주어졌을 때
n ^ m 을 구하는 프로그램을 작성합니다.

물론 파이썬 에서는 ** 연산자를 통해 쉽게 연산이 가능하지만,

연습을 위해서 직접 만들어 보겠습니다.

임의의 정수는 5 이하의 정수라고 가정합니다.

n ^ m = n(1) * n(2) * n(3) * ... * n(m)

2 ^ 3 = 2 * 2 * 2

5 ^ 4 = 5 * 5 * 5 * 5

```python
n = 5
m = 1
for i in range(1, n) :
  m *= n

  print('{} ^ {} = {}'.format(n, i, m))
```

```python
def oops(a, b) :
  ans = 1
  for i in range(1, b+1) :
    ans *= a
  return ans
```

```python
oops(5, 4)
```
