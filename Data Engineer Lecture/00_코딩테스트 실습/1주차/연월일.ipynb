```
출력
#1 2222/02/28
#2 -1
#3 0101/01/01
#4 -1
#5 1111/11/11
```

````python
a = int(input())

for i in range(1, a+1) :

  b = str(input())
  if int(b[4:6]) in [1,3,5,7,8,10,12] :
    if int(b[6:8])>=1 and int(b[6:8])<=31 :
      print(f'# {i} {b[0:4]}/{b[4:6]}/{b[6:8]}')
    else :
      print(f'# {i} {-1}')

  elif int(b[4:6]) in [4,6,9,11] :
    if int(b[6:8])>=1 and int(b[6:8])<=30 :
      print(f'# {i} {b[0:4]}/{b[4:6]}/{b[6:8]}')
    else :
      print(f'# {i} {-1}')  

  elif int(b[4:6]) in [2] :
    if int(b[6:8])>=1 and int(b[6:8])<=28 :
      print(f'# {i} {b[0:4]}/{b[4:6]}/{b[6:8]}')
    else :
      print(f'# {i} {-1}')  

  else :
    print(f'# {i} {-1}')  
````
