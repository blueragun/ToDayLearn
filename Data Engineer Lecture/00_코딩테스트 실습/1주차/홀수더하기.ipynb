```python
# 정답

a = int(input())

for i in range(1, a+1) :
    ans = 0
    for j in list(map(int, input().split())) :
      if (j%2 == 1) :
          ans += j
    print(f'#{i} {ans}')
```

# 별찍기...확장판이다.... 하.. 씁

```python
# 오답

a = int(input())
b = list(map(int, input().split()))
ans = 0

for i in range(1, a+1) :
    for j in b :
      if (j%2 == 1) :
          ans += j
    print(f'#{i} {ans}')

# ans(총합) 이 바깥에 있기 때문에 1차 for 계산 결과값 200이 그대로 남았고
# 중첩 for 문으로 인해 2차, 3차가 그대로 계산됨.....
```

```python
# 오답

a = int(input())
b = list(map(int, input().split()))

for i in range(1, a+1) :
    ans = 0
    for j in b :
      if (j%2 == 1) :
          ans += j
    print(f'#{i} {ans}')

# 1번 for문에 ans를 넣으니 하위 for문에서 계산이 종료되면 0으로 되어 재계산됨
# 다만 아직도 추가 input을 받지 않고....상위 포문 그대로 계산...쩝
```
