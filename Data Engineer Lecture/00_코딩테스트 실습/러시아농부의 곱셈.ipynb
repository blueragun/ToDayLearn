# 러시아농부의 곱셈
옵션
왼쪽에 있는 수는 2로 나눠주고, 나머지는 버린다. (1이 될때까지 반복)

오른쪽에 있는 수는 두배를 해준다.
왼쪽에 있는 수가 홀수인 오른쪽의 수들을 모두 더해준다.

123 x 12
61 x 24
30 x 48
15 x 96
7 x 192
3 x 384
1 x 768

```python
a = 123
b = 12
russia = 0

while (a >= 1) :
  if(a%2 != 0) :
    russia += b
    b *= 2
    a /= 2

  if(a%2 == 0) :
    b *= 2
    a /= 2

print(russia)

# a X b를 할때 a는 2로 나눠준다 / 나머지는 버리고 1이 될때 까지 반복한다.#
# b는 두배를 곱해준다.
```
