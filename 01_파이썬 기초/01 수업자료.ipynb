# QnA
1. 맥북이나 맥os를 사용해도 수업에는 지장이 없는지? 
  - 전혀 상관은 없습니다. 
  - 보통 실습은 클라우드에서 진행(로컬 환경을 잘 사용하지 않습니다)
    - 파이썬 기초, 데이터 시각화, 통계분석, ... 
    - 풀스택 => 로컬 환경에서 진행
      - 개인 컴퓨터에 개발 환경을 설치해서 진행
      - 윈도우즈와 맥OS와 차이가 발생(사용하는 툴은 동일)
    - 엔지니어링 수업 => AWS(원격의 리눅스 시스템)
  - 수업은 윈도우즈 기반에서 진행

2. 컴퓨터 사양은 어느정도?
  - 노트북을 구매하고 싶은데 사양은 어느정도가 좋을까요?
  - 맥북을 구매해도 괜찮을 까요?
  - 노트북을 구매할 예정이시라면
    - 여러분이 좋아하는, 즐겨하는 게임의 사양에 맞춰주시면 됩니다.
    - 일반적인 데스크탑이나, 노트북은 분석 환경에 맞추기 힘듭니다.
    - 개인이 구비하기에는 너무 비싸요
  - 맥북이나 아이맥, 등을 고려하신다면
    - 분석용으로는 맥은 비추 합니다. 
    - 맥은 예쁘면 구입하세요
    - 가장 유명한 딥러링 라이브러리인 `텐서플로우`가 NVIDIA 그래픽 카드가 아니면 사용할 수 없습니다.

3. SQL은 기초적인 DML을 소개하는 정도로 다룰 예정입니다. 
  - 파이썬에서는 SQL을 대체해서 사용할 수 있는 라이브러리가 있습니다. 
  - 파이썬 장고 프레임워크에서는 SQL을 몰라도 DB를 사용할 수 있습니다. 

4. 추천할만한 책은 없는지? 
  - 교재가 따로 있기 때문에 교재를 충분히 활용하시면 제일 좋을 것 같습니다. 
  - 추천 도서
    - 수업과는 전혀 상관이 없는 교양서적 입니다. 
    - 하지만! 코린이 여러분들은 꼭! 읽어보셨으면 좋겠습니다. 
    - [컴퓨터 과학이 여는 세계](http://www.kyobobook.co.kr/product/detailViewKor.laf?ejkGb=KOR&mallGb=KOR&barcode=9788966261437&orderClick=LEa&Kc=)
      - 서울대학교에서 교양과목으로 진행되는 수업을 책으로 만든겁니다. 
      - 전공자를 대상으로 하는 책이 아니라는 말입니다. 
      - 특히, 컴퓨팅 환경에 익숙하지 않은 분들은 필독 하셨으면 좋겠습니다

  - 파이썬을 너무 어려워 하거나, 코딩 자체를 어려워 하는 경우가 많습니다. 
    - 당연하지만, 프로그래밍 언어를 처음 접하면 굉장히 어려워요
    - 대부분의 경우 컴퓨팅 환경에 익숙하지 않아서 발생하지 않나 생각합니다. 

  - 우리가 코딩을 어려워 하는 이유?
    - `식빵에 잼을 바르는것과 동일한 이유`가 되지 않을까 합니다. 
    - 우리가 일반적으로 생각하는 식빵에 잼을 바르는 과정과 컴퓨터가 생각하는 과정은 달라요
    - 언젠가는 누구나다 이해하는 내용이지만, 처음에는 굉장히 어려워요
    - 익숙해질때까지는 어려움이 있을수 밖에 없고, 이 수업은 반드시 한번은 거쳐가야 되는 과정이에요
      - 식빵에 잼을 바르는 방법(과정) => 알고리즘
      - 식빵이 10만개가 되면, 관리해야 하는데 => 자료구조

  - 여러분 수업 열심히 듣지 마세요, 그냥 들으시고, 이해안되면 그냥 넘어가세요 
    - 수업이 스트레스면 오래 못가요, 꾸준하게 오래 갈 수 있으면 좋겠어요 
    - 이 수업이 끝이 아니고, 내일도 수업이 있고, 다음주에도 수업이 있고, 다음달에도 수업이 있어요
    - 취업해도 끝이 아니에요, 커리어도 쌓고, 공부를 계속 지속
    - 이 수업 하나로 다 끝내려고 하지 마시고

  - 저는 대놓고 지루하게 수업을 합니다. 
  - 코린이 여러분들을 대상으로 진행을 합니다. 
  - 전공자 분들이나, 이미 해당 과정에 대한 이해도가 어느정도 있으신 분들은 지루할 수 있습니다. 
    - 너른 양해를 부탁드리겠습니다.
