1. 소인수분해 대상 input() 받고
2. 결과값 0으로 설정하고
3. while로 1이 아닐때 까지 반복하게 설정

```python
T = int(input())

for i in range(1, T+1) :

  m = int(input())
  a = 0; b = 0; c = 0; d = 0; e = 0;
  while(m != 1) :
    if m % 2 == 0:
      m /= 2
      a += 1
    elif m % 3 == 0:
      m /= 3
      b += 1
    elif m % 5 == 0:
      m /= 5
      c += 1
    elif m % 7 == 0:
      m /= 7
      d += 1
    elif m % 11 == 0:
      m /= 11
      e += 1

  print(f'#{i} {a} {b} {c} {d} {e}')
```
