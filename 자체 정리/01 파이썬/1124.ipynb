# 시퀀스 타입
- 기본 타입(숫자, 문자열)으로 이루어진 여러개의 자료를 하나의 자료형으로 다루기 
  위한 방법
  * 다른 언어에서는 '배열'이라고 표현
  * 한개의 식빵이 기본 타입이라고 한다면, 여러개의 식빵을 다루기 위한 자료의 타입 중에 하나로 보면 됨
- 파이썬의 시퀀스 타입은 가장 기본적인 '자료구조'의 한 종류
  * 자료구조는 여러개의 자료를 관리하는 방법
- 파이썬에서 제공하는 시퀀스 타입의 종류
  * 리스트
    - 다른 언어에서 제공하는 '배열'의 개념이 파이썬에서는 리스트이다.
    - 가장 많은 사용하는 자료형
    - '선형자료구조'라고 한다.
  * 튜플
    - 리스트와 동일한데, 가장 큰 차이는 자료의 값을 변경할 수 있느냐 없느냐의 차
    - 이런 파이를 제외하고는 리스트와 동일
    - 튜플은 튜플내의 자료들의 값이 변하지 않음을 보장한다.
  * 레인지
    - 수열을 생성

# 매핑형 타입
- 파이썬에서는 '딕셔너리'
- 다른 환경에서는 '해시 자료구조'

- 딕셔너리 형태는
  * { key:value, key:value, ...}


- 키를 직접 정의 가능
  * 튜플이나 리스트와는 다르게 인덱스가 자동생성되지 않음
  * 반드시 숫자가 아니어도 됨
  * 즉, 인덱스 값 지정 가능

```python
# 빈 딕셔너리를 만드는 방법
hash = {}
type(hash)
```

```python
# 딕셔너리인 경우 '키'가 존재하면 값을 업데이트 하고
# 키가 존재하지 않으면, 자동으로 추가
# 리스트처럼 원소를 추가하기 위해서 명령어를 사용하지 않음
hash['first'] = 10
```

```python
hash
```

```python
# 자동으로 인덱스가 생성되지 않음
# 예제
hash[0]
```

```python
# 다음과 같이 참조 가능
hash['first']
```

```python
# 키가 존재하면 값을 업데이트하게 된다
hash['first'] = 20
hash
```

```python
# 키가 존재하지 않으면, 새로운 키를 추가하게 됨
hash['second'] = 30
hash
```

# 딕셔너리는 가변하기도 하면서, 불가변하기도 함
- 키는 수정할 수 없다
- 값은 수정할 수 있다

# 키가 해시값
