# 함수의 호출 구조
1. 함수가 언제 실행 되는가? 
  - 함수는 호출고 함께 실행된다
  - 함수가 실행 된다는 것은?
    * **메모리에 함수를 위한 공간이 따로 만들어진다는 것**
    * '콜스텍(호출스텍' 이라 한다
    * 즉, 함수는 함수가 사용하는 고유 메모리 공간을 따로 가지게 된다.

2. 함수는 언제 종료가 되는가?
  - 함수를 블록 내에서 더 이상 실행할 명령어가 없으면 종료
  - 또는 'return'명령어가 실행되면 종료
  - 함수가 종료된다는 것은?
    * **함수가 사용했던 메모리는 존재 안함**  
    * 함수는 종료하면서 사용했던 메모리를 반납하게 된다

3. continue, break, return, yield
  - continue, break는 반복문에서만 사용
  - return, yield 함수에서만 사용

# 재귀함수
  - 자기자신을 호출하는 함수를 재귀함수 혹은 재귀호출 이라고 함
```
def recursive():
    print('무한히 반복하는 형태')
    rerursive()     

rerursice()  //트리거
```

````python
# 재귀는 반복의 또 다른 형태
# while, for를 사용하지 않고도 반복적인 처리 가능
# 잘못쓰면, 무한히 재귀 호출이 발생(시스템이 다운될 가능성이 있음)

def recursive(n) :
  # 기저조건(함수를 종료시켜 주는 조건문)
  if n > 10: return

  print("{} 재귀함수가 뭔가요?" .format('--' *n))
  print("{} 잘 들어보게.".format('--' *n))

  recursive(n+2)
````

1부터 n까지의 합을 구하는 프로그램
- 일반화된 공식도 있다
  * 가우스가 발견
  * 반복적으로 처리

````python
# 반복문을 이용해서 1부터 n까지의 전체 합
# 1 + 2 + 3 + ....+ n
n = 10
````

````python
# for를 이용한다면 더해줘야 되는 수를 range로 생성하는 방법을 생각해복 수 있다
ans = 0
for i in range(1, n+1):
  ans += i
print(ans)
````

````python
ans = 0
while n:
  ans += n
  n -= 1
print(ans)
````

````python
sum = 1
for i in range(11) :
  sum += i
print(sum)
````

# 익명함수
  - 파이썬에서 람다를 이용해서 익명함수를 작성할 수 있다
  - 함수의 이름이 없기 때문에 익명함수라고 한다
    * 자주 사용되는 기능은 아니지만 은근 쓰임
    * 함수 이름이 없으면 어떻게 호출/??
      - 두고두고 사용하는 그런 함수의 개념과 다름


    ```
    lamda 파라미터 : 명령어
    ```
    
    - 여러줄에 걸친 복잡한 명령은 사용불가
      * 한 라인에 정의 가능한(간단한) 형태를 사용할 떄 유용

````python
(lambda a, b: a + b)(10, 20)
````

````python
def add(a, b):
  return(a + b)

add(10, 20)
````

````python
print((lambda a, b : a + b)(10, 20))
````

# 함수형 문법
  - 파이썬은 거의 대부분의 문법을 지원
  1. 절차(프로시져)형 프로그래밍
    - 프로시져가 국내로 들어오면서 '절차'라는 단어로 해석
    - '부프로그램'의 의미로 해석  => 함수
    - 소스코드를 함수 단위로 관리
      * 소스코드의 재사용성이 증가 => 소스코드도 양이 줄어드는 효과
      * 관리 용이
    - 파이썬은 함수만 이용해서 코드를 작성할 수 있다

  2. 객체지향
    - 함수 단위로 소스코드를 작성해도 해결이 안됨
      * 새로운 개발 방법론이 필요
      * 클래스라는 개념이 도입
        - 함수를 사용했을 때보다 재사용성이 더 증가
        - 완벽하게 깨끗한 SW 만들어 내는 것은 불가능

  3. 함수형 프로그래밍
    - 코드를 작성하는 완벽히 새로운 방법
    - 원래도 존재했던 이론이지만, 현대에 들어서면서 점점 많은 사람들에게 선택되어지기 시작
      * 사유 : 완벽하지 않은 SW

````python
# 파이썬의 익명함수를 응용해보면
# f(x) = 2x와 같은 함수를 가정

x = [1, 2, 3, 4]
list(map(lambda x : 2 * x, x))
````
