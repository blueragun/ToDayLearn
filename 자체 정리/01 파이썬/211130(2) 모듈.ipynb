# 예외처리
- 코드를 작성하면서 발생할 수 있는 예외(에러, 경고, .. )에 대해서 처리할 수 있는 방법 2가지
  1. LBYL(Look Before You Leap)
    - `도약하기 전에 확인해라`
    - 코드를 실행하기 전에 예외가 발생하지 않도록 미리 검증하고 실행을 하자
    - 즉, 발생할 수 있는 모든 예외적인 상황들에 대해서 미리 예측하고, 미연에 방지 
    - 현실적으로는 불가능한 방법
      - 완벽하게 예외를 예측하고, 전부 처리할 수 없습니다. 
      - 만약에 이게 가능했다면, 패치나 업데이트 라는 개념도 없었겠죠 
    - 전통적으로 예외를 처리해오던 방법

  2. EAFP(It's Easier Ask Forgiveness Than Permission)
    - `허락보다 용서가 쉽다`
      - 모 광고의 광고문구로도 사용이 되었습니다.(플스5)
    - 일단 코드를 실행하고, 예외가 발생하면, 그때 처리하자
    - 파이썬 표준에서 권장하는 방식

## 전통적인 방식의 예외처리

````python
# 문자열을 입력하면, 숫자로 바꿔서 돌려주는 함수를 가정
def convert(strs):
  return int(strs)
````

````python
print( int('10') )
print( int('-10') )
print( int('+10') )
````

````python
# 전통적인 방식으로 이러한 예외를 처리한다면
# 함수 내에서는 숫자로 변경 가능한 경우에만 변환할 수 있도록 체크
def convert(strs):
  
  # 1. 입력이 반드시 문자열 일거라는 보장이 있는가?
  # 그래서 타입 체크를 먼저 수행(문자열인 결우에만 진행 가능)
  if isinstance(strs, str):
  
  # 2. 문자열이라면, 첫번째 원소의 값은 반드시 셋 중에 하나(-, +, 정수)가 되어야 한다.
    if strs[0] == '-' or strs[0] == '+' or strs[0].isdecimal():
  
  # 3. 첫번째 원소를 제외한 나머지 원소는 전부 정수여야 한다.
      if strs[1:].isdecimal():
        return int(strs)
    
      else: return False

    else: return False

  else: return False
````

# EAFP 방식의 예외처리

````python
def convert(strs):
  # 1. 예외가 발생할 것 같은 코드를 try 블록으로 넣어준다

  try:
    return int(strs)
  except: # try 블록내의 코드에서 예외가 발생하면 except 블록으로 예외가 넘어오게 된다.
    return False
````

# 다양한 형태의 예외처리 방법

# 특정 예외만 처리하는 경우


````python
def convert(strs):
  try:
    return int(strs)
  except ValueError:   # ValueError는 클래스 즉! 객체임
    return False
````

````python
convert('a')
````

````python
convert(10)
````

`elif` 구문처럼 여러개의 예외를 처리할 수 있다

````python
def convert(strs):
  try:
    return int(strs)
  except ValueError:   # ValueError는 클래스 즉! 객체임
    print('valueError')
  except TypeError:  # elif처럼 여러개 예외 가능
    print('TypeError')
````

````python
convert('a')
````

````python
convert([0])
````

# 예외별로 정확한 예외 메시지를 확인

````python
def convert(strs):
  try:
    return int(strs)
  except ValueError as e:   # ValueError는 클래스 즉! 객체임
    print(e)
  except TypeError as e :  # elif처럼 여러개 예외 가능
    print(e)
````

````python
convert('a')
````

````python
convert([])
````

````python
def convert(strs):
  try:
    return int(strs)

  # 주의!!
  # 가장 큰 범위의 예외 클래스가 먼저 오지 않도록 해야함
  except Exception as e:  
    print('언노운 엑셉션', e)
  except ValueError as e:
    print('발유에러', e)
````

### 확장된 형태
- else, finally

````python
def convert(strs):
  try:
    print( int(strs) )
  except ValueError as e:
    print('발유에러', e)

  except Exception as e:  
    print('언노운 엑셉션', e)
  
  else:
    print('예외가 발생하지 않으면 실행')
  
  finally:
    print('예외 유무와 상관없이 마지막에 항상 무조건 실행')
````

````python
convert('10')
````

````python
convert('a')
````

````python
while True :
# 표준 입력을 받는 함수
# input은 키보드로 한 라인을 입력
  read = input()
  print( read )

  # ctrl + d 하면 에러문구가 뜨면서 입력 종료
````

### 처리된지 않은 예외
- 모든 예외를 전부 처리할 수 없다
- 예외도 여러 종류가 있다
  - 에러
    1. 컴파일 에러
      - 파이썬은 컴파일 과정은 없고 인터프리터 방식으로 동작
      - 컴파일이 아닌 인터프리터 과정에서 발생하는 애러
      - REPL 과정에서 READ => Evaluaion 사이에 번역이 이루어 지는데, 그때 발생하는 에러
      - 문법적인 오류등은 이 과정에서 발생
    2. 런타임 에러
      - 프로그램을 실행 중(프로세스)인 상태에서 발생하는 에러
      - 번역은 끝난 상태에서(문법적인 오류는 없는 상태) 실행하는 도중에 문제가 발생
      - 파이썬 Evaluaton 과정에서 발생하는 오류
  - 경고 메시지
  - 에러는 여러가지 예외들 중에서도 프로그램 실행에 영향을 줄 만큼 심각한 오류
