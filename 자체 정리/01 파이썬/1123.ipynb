```python
# 타입변환
- 파이썬은 비교적 자유롭게 타입을 변환할 수 있다.
- str(), int(), float(), bool()
- str(x) : x를 문자열로 변환한 결과를 반환해준다

- int(x) : 실수, 문자열 x에 대해 정수로 변환한 결과를 반환해준다

- float(x) : 정수, 문자열 x에 대해 실수로 변환한 결과를 반환해준다

- bool(x) : 모든 타입의 x에 대해 전부 참 또는 거짓으로 변환된 값을 반환해준다
   - 거짓이 되는 경우를 제외하면 모두 참임
```
