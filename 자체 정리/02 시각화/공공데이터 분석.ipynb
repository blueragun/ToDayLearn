# 앞에서 했던 작업을 함수로

## 함수로 정의
- 전처리 작업을 5개의 모든 파일에 동일하게 적용
- DF를 매개변수로 받아서 전처리가 완료된 후에 데이터 프레임을 반환하는 함수

```python
def conv(x):
  x.fillna('0', inplace=True)  # nan 값을 0으로 처리시킴
  if x.name == '행정구역': return x
  return x.str.replace(',', '').astype(np.int)  # 문자열 중 ','를 없애고 int 타입으로 변환

def preProcessing( df ):  # 함수를 이용해 기존에 했던 작업을 종합
  tmp = df.iloc[:, 0::3]
  tmp['행정구역'] = tmp['행정구역'].str.split(' ').str[0]   # 문자열 인덱스 0번째의 ' ' 후의 문자 제거
  names = []     # 행정구역의 문구 포멧을 변경
  for name in tmp.columns:
    if name == '행정구역':
      names.append(name)
      continue
    names.append(pd.to_datetime(name.split('_')[0], format='%Y년%m월'))
  tmp.columns = names
  tmp = tmp.apply( conv )
  tmp = tmp.set_index('행정구역')

  return tmp.T
```

### 나머지 파일을 모두에 대해 작업내용 반영

```python
tmp1 = pd.read_csv('위치', encoding='cp949')
tmp1 = preProcessing(tmp1)
```

```python
tmp2 = pd.read_csv('위치', encoding='cp949')
tmp2 = preProcessing(tmp2)
```

```python
tmp3 = pd.read_csv('위치', encoding='cp949')
tmp3 = preProcessing(tmp3)
```

```python
tmp4 = pd.read_csv('위치', encoding='cp949')
tmp4 = preProcessing(tmp4
```

```python
tmp5 = pd.read_csv('위치', encoding='cp949')
tmp5 = preProcessing(tmp5)
```

## concat을 이용한 데이터 프레임 합치기

```python
rawData = pd.concat( [tmp1, tmp2] )
rawData = pd.concat( [rawData, tmp3] )
rawData = pd.concat( [rawData, tmp4] )
rawData = pd.concat( [rawData, tmp5] )
rawData
```
