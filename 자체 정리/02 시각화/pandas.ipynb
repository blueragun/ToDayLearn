## apply
- 자료를 여러분이 원하는데로 다루기는 쉽지 않다
  - 실제로 데이터를 다뤄보면 내 생각처럼 쉽게 다뤄지지 않는다.
  - 상당히 어렵다

- 판다스에서 사용가능한 반복문 정도로 이해
- 모든 행, 열에 대해서 동일한 함수를 반복적으로 적용

````python
# apply의 기본적인 사용방법
def func(x) :
  print('함수 호출')
  print('x:{}'.format(x))

# 기본적으로는 데이터 프레임의 각 컬럼들이 시리즈 형태로 반환된다
rawData.apply( func )
````

````python
def func(x) :
  print('함수 호출')
  print('x:{}'.format(x))
  # none을 방지하려면 return을 반드시 작성
  return()

rawData.apply( func )
````

````python
# 칼럼이 아닌, 행을 전달하고 싶다면
def func(x) :
  print('함수 호출')
  print('x:{}'.format(x))

  return

rawData.apply( func, axis=1 )
````

## 프레임 합치기
- 서로다른 두 자료를 하나의 데이터 프레임으로 합치는 방법
- 여러 소스(출처)부터 가져온 데이터를 분석하기 위해서는 반드시 하나의 데이터 프레임이어야만 함



````python
user_device = pd.read_csv('/content/drive/MyDrive/데이터교육/수업자료/user_device.csv')
user_usage = pd.read_csv('/content/drive/MyDrive/데이터교육/수업자료/user_usage.csv')
````

## inner join
- merge 함수를 통해서 `join`을 해볼 수 있다
  - merge의 기본동작은 `inner join`이다

- 이너 조인은 기본적으로 자료의 크기가 훨씬 줄어들게 된다
  - user_usage : 240개
  - user_device : 270개
    - 두 자료를 하나로 합쳤을 경우 240개를 넘어가지 않는다
    - 양쪽 자료 모두에 존재하지 않는 자료가 있다면?

````python
display( len(user_device) )
display( len(user_usage))
````

````python
# 공통으로 존재하는 컬럼을 기준으로 이너 조인
pd.merge(left=user_usage, right=user_device, on='use_id')
````

## left join
- 왼쪽 자료를 기준으로 합쳐준다
- 결측치가 발생할 수 있다
  - 오른쪽에 없는 자료는 결측치로 채워지게 된다

- 왼쪽에는 있지만 오른쪽에 없는 자료는 결측치로 채움
- 행 갯수는 왼쪽을 기준으로 동일하게 유지

````python
pd.merge(left=user_usage, right=user_device, on='use_id', how='left')
````

## right join
- 오른쪽 자료를 기준으로 합쳐준다
- 결측치가 발생할 수 있다
  - 왼쪽에 없는 자료는 결측치로 채워지게 된다

- 오른쪽에는 있지만 왼쪽에 없는 자료는 결측치로 채움
- 행 갯수는 오른쪽을 기준으로 동일하게 유지

````python
pd.merge(left=user_usage, right=user_device, on='use_id', how='right')
````

## outer join
- 왼쪽, 오른쪽 자료가 모두 출력  (+)
- 양쪽에 없는 데이터는 결측치 처리

````python
pd.merge(left=user_usage, right=user_device, on='use_id', how='outer')
````
