# 시각화
- 자료 분석의 기본은 통계와 그래프 입니다.

- 시각화 하는 방법은 정해져 있습니다.
  - 자료(변수)의 형태에 따라서, 그릴 수 있는 그래프는 이미 결정
  - 자료(변수)의 개수에 따라서 그릴 수  있는 그래프는 이미 결정

    - 전통적인 통계적인 분석에서 3변수 이상은 잘 다루지 않는다
    - 현재 빅 데이터는 아주아주 많은 변수를 다룬다고 보면 된다

# 그래프를 그리는 기준
- 통계분석에서도 동일한 기준으로 분석 함

1. 자료 형태(변수의 종류)
  - 범주형(문자)과 연속형(숫자) 자료   / 단. 명확히 구분되지 않음
  - 질적변수 vs 양적변수 로 나눔

  - 질적변수(Qualitative variable) : 변수의 값이 비수치적, 특정 카테고리에 포함 되어있는 경우
    - 명목형 : 특정범주에 포함되지만, 순위를 갖지 않음 / ex) 성별, 혈액형 등
    - 순위형 : 특정범주에 포함되지만, 순위를 가지 / ex) 성적, 등급 등

  - 양적변수(Quantative variable) : 변수의 값이 수치적, 연속적(키, 나이 등)
    - 이산형 : 셀 수 있는 경우(정수인 경우)
    - 연속형 : 셀 수 없는 경우(실수인 경우)


2. 자료의 차원
  - 변수의 갯수는 차원이 됨 / 변수가 많아질수록 차원이 높아진다
    - 차원이 높으면 시각화를 할 수 없다
      - 최대 2차원의 화면에서 3차원까지만 나타낼 수 있음
  - 1차원 
    - 일 변수(일변량) : 변수 하나에 대한 시각화
    - 변수의 분포 등을 확인
  - 2차원
    - 이 변수(이변량) : 변수 - 변수 간의 관계를 확인하고 싶은 경우
      - 연속형 변수 - 연속형 변수
      - 연속형 변수 - 이산형 변수
      - 이산형 변수 - 이산형 변수

#### boxplot
- 탐색적 자료 분석(EDA)에서 주로 많이 사용
- 4분위수를 이용한 시각화
  - 이상치(outlier)를 확인하는 용도로 사용
  - 자료를 4개의 구간으로 나눈 값
  - 1분위수: 자료의 25%에 해당하는 값
  - 2분위수: 자료의 50%에 해당하는 값(중앙값과 동일)
  - 3분위수: 자료의 75%에 해당하는 값

![](https://i.imgur.com/tssRkgA.jpg)

- IQR(Inter Quantile Range)
  - IQR = Q3 - Q1
  - 최대제한선 = Q3 + (IQR * 1.5)
  - 최소제한선 = Q1 - (IQR * 1.5)
